```python
from colomoto.minibn import *
```

Example of section 1.4.1.1:

```python
f = BooleanNetwork({
    "x1": "x3",
    "x2": "x1",
    "x3": "x2"
})
f.influence_graph()
```

```python
ds = SynchronousDynamics(f, loops=True).dynamics()
ds
```

```python
{x for (x,y) in ds.edges() if x == y}
```

```python
dp = PeriodicDynamics(({"x2", "x3"}, {"x1", "x3"}, {"x1", "x2"}), f).dynamics()
dp
```

```python
{x for (x,y) in dp.edges() if x == y}
```
