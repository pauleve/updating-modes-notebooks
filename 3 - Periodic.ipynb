# Periodic updating modes

The `minibn.PeriodicDynamics` allows specifying any finite periodic updating mode, by giving at first argument the sequence of parallel updates, exactly as with `BlockSequentialDynamics`.

```python
from colomoto.minibn import *
```

Let us consider the Boolean network example of section 6.4.1.1:

```python
f = BooleanNetwork({
    "x1": "x3",
    "x2": "x1",
    "x3": "x2"
})
f.influence_graph()
```

Its synchronous (parallel) dynamics is the following:

```python
ds = SynchronousDynamics(f).dynamics()
ds
```

and it contains 2 fixed points:

```python
{x for (x,y) in ds.edges() if x == y}
```

Now, let us consider the following periodic dynamics (which is not a partition of the set of automata):

```python
pd = PeriodicDynamics(({"x2", "x3"}, {"x1", "x3"}, {"x1", "x2"}), f).dynamics()
pd
```

This periodic updating contains two additional fixed points:

```python
{x for (x,y) in pd.edges() if x == y}
```
