# Block-parallel updating modes

This notebook shows how to reproduce parts of Figure 6.4 and 6.6 using the `minibn.BlockParallelDynamics` class.

```python
from colomoto.minibn import *
```

```python
f = BooleanNetwork({
    "x1": "!x3",
    "x2": "x2 & (x1 | x3)",
    "x3": "!x1"
})
f
```

Similarly to the `BlockSequentialDynamics` class, a `BlockParallelDynamics` instance takes as a first argument the partitioned order of the automata: each element specifies a sequence of automata.
The sequences are given as standard Python `list` or `tuple`, or a single automaton name.
These elements are enclosed in a `list`, `tuple`, or `set` object.

The following code computes the top-left graph of Figure 6.4 (three blocks of single automata):

```python
BlockParallelDynamics({"x1", "x2", "x3"}, f).dynamics()
```

The following code computes the top-middle graph of Figure 6.4:

```python
BlockParallelDynamics({"x2", ("x1","x3")}, f).dynamics()
```

The `BlockParallelDynamics` also support *waiting delays*, specifed by `None`. The following code computes the right graph of Figure 6.6:

```python
BlockParallelDynamics({("x1","x2"), ("x3",None)}, f).dynamics()
```

The following code computes the middle graph of Figure 6.6:

```python
BlockParallelDynamics({("x1",None,"x2"), ("x3",None)}, f).dynamics()
```
