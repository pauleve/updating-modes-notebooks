## Manual input of Boolean networks

Boolean networks can be specified directly using the `BooleanNetwork` class. It consists of a dictionnary mapping variable names to their local function. Negation symbol is `!`, conjunction is `&`, and disjunction is `|`.

From Figure 1.1(a):

```python
f = BooleanNetwork({
    "x1": "!x3",
    "x2": "x2 & (x1 | x3)",
    "x3": "!x1"
})
f
```

```python
f.influence_graph()
```

From Figure 1.1 (b):

```python
g = BooleanNetwork({
    "x1": "x1 | !x2 | !x3",
    "x2": "x3 | (x1&!x2) | (!x1&x2)",
    "x3": "!x1 | !x2 | !x3"
})
g
```

```python
g.influence_graph()
```
