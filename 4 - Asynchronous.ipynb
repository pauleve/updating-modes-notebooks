# Asynchronous updating modes

The `GeneralAsynchronousDynamics` and `FullyAsynchronousDynamics` have no parameters and compute respectively the (general) asynchronous and the fully asynchronous updating mode. By default, loops are omitted.

This notebook shows how to compute the graphs of Figure 1.9.

```python
from colomoto.minibn import *
```

```python
f = BooleanNetwork({
    "x1": "!x3",
    "x2": "x2 & (x1 | x3)",
    "x3": "!x1"
})
f
```

```python
GeneralAsynchronousDynamics(f).dynamics()
```

```python
FullyAsynchronousDynamics(f).dynamics()
```

Loops can be included using the `loops=True` option:

```python
FullyAsynchronousDynamics(f, loops=True).dynamics()
```
